```python
from scipy.sparse import csc_matrix 
from scipy.sparse import random 
```

```python
def get_sparse_size(matrix):
    return int((matrix.data.nbytes + matrix.indptr.nbytes + matrix.indices.nbytes)/1024.) #get the size of the matrix 
sparse_mat = random(10**3, 10**5, format = 'csr') #create a sparse matrix of 1000 x 100000
sparse_size = get_sparse_size(sparse_mat) #get the size of sparse matrix 

#convert the sparse matrix into regular matrix and get its size 
regular_size = sparse_mat.toarray().nbytes/1024.
print("The size of the sparse matrix is %s KiB" % sparse_size)
print("The size of regular matrix is %s KiB" % regular_size)
print("Data compression ratio is %s" % (regular_size/sparse_size))
```

```python
import pandas as pd
import matplotlib.pyplot as plt


url = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
#load dataset into Pandas DataFrame
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])

print(df.head(1))
```

```python
from sklearn.preprocessing import StandardScaler
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
# Separating out the features
x = df.loc[:, features].values
# Separating out the target
y = df.loc[:,['target']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
print(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])

print(principalDf.head())
```

```python

finalDf = pd.concat([principalDf, df[['target']]], axis = 1)

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
plt.show()
```

```python
print(pca.explained_variance_ratio_)
```

```python
from pandas import Series
from pandas import DataFrame
from pandas import concat
series = Series.from_csv('C:/Users/GQPF6681/Desktop/math-for-mlv1-master/feature-engg/daily-temp.csv', header=0)
temps = DataFrame(series.values)

print(temps.head())
```

```python
#Sliding window of size 1
print('Sliding window of size 1')
dataframe = concat([temps.shift(1), temps], axis=1)
dataframe.columns = ['t-1', 't+1']
print(dataframe.head(5))
```

```python
#Sliding window of size 3
print('Sliding window of size 2')
dataframe = concat([temps.shift(2), temps.shift(1), temps], axis=1)
dataframe.columns = ['t-2', 't-1', 't']
print(dataframe.head(5))
```

```python
#Sliding window of size 3
print('Sliding window of size 3')
dataframe = concat([temps.shift(3), temps.shift(2), temps.shift(1), temps], axis=1)
dataframe.columns = ['t-3', 't-2', 't-1', 't']
print(dataframe.head(5))
```

```python
shifted = temps.shift(1)
window = shifted.rolling(window=2)
print('++++++++++++++++')
print(window)
means = window.mean()
dataframe = concat([means, temps], axis=1)
dataframe.columns = ['mean(t-2,t-1)', 't+1']
print(dataframe.head(5))
```

```python
#More statistics on the rolling window
print('Adding additional statistucs')
width = 3
shifted = temps.shift(width - 1)
window = shifted.rolling(window=width)
dataframe = concat([window.min(), window.mean(), window.max(), temps], axis=1)
dataframe.columns = ['min', 'mean', 'max', 't+1']
print(dataframe.head(5))
```
